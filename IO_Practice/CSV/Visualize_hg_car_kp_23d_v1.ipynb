# Visualization

## Loss

```python
# Overall view
plt.plot(df_train_loss.iter_count.iloc[:-1], df_train_loss.loss[:-1])
plt.title("loss vs. iter_count 0 - 350k")
plt.xlabel('iter_count')
plt.ylabel('CE loss')
```

```python
# zoom in
plt.plot(df_train_loss.iter_count.iloc[100:-1], df_train_loss.loss[100:-1])
plt.title("loss vs. iter_count 0 - 350k")
plt.xlabel('iter_count')
plt.ylabel('CE loss')
```

```python
# moving average
win_size = 30
df_train_loss['MA'] = df_train_loss.loss.rolling(window=win_size).mean()

plt.plot(df_train_loss.iter_count.iloc[100:-win_size+1], df_train_loss.MA[100:-win_size+1])
plt.title("loss vs. iter_count MVA smoothed with Win_size=30")
plt.xlabel('iter_count')
plt.ylabel('CE loss')
```

## Accuracy on Validation Set - C0 (full image)

```python
plt.plot(df_val_loss_avg_accur.steps, df_val_loss_avg_accur.c0_avg_accur)
plt.title("Class 0 Avg Accuracy vs. iter_count 0 - 350k")
plt.xlabel('iter_count')
plt.ylabel('Accuracdy')
```

```python
win_size = 5
df_val_loss_avg_accur['MA0'] = df_val_loss_avg_accur.c0_avg_accur.rolling(window=win_size).mean()

plt.plot(df_val_loss_avg_accur.steps.iloc[win_size-1:-win_size+1], df_val_loss_avg_accur.MA0[win_size-1:-win_size+1])
plt.title("C0 Avg Accuracy vs. iter_count 0 - 350k")
plt.xlabel('iter_count')
plt.ylabel('Accuracdy - Moving Avg (window=5)')
```
