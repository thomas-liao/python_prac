# Loss

```python
# Overall view
plt.plot(df_train_loss.iter_count.iloc[:-1], df_train_loss.loss[:-1])
plt.title("loss vs. iter_count 0 - 200000")
plt.xlabel('iter_count')
plt.ylabel('CE loss')
```

```python
# zoom in
# Overall view
win_size = 30
df_train_loss['MA'] = df_train_loss.loss.rolling(window=win_size).mean()

plt.plot(df_train_loss.iter_count.iloc[100:-win_size+1], df_train_loss.MA[100:-win_size+1])
plt.title("loss vs. iter_count with Win_size=30")
plt.xlabel('iter_count')
plt.ylabel('CE loss')
```

# Accuracy - C0

```python
win_size = 5
start = 20
df_val_loss_avg_accur['MA0'] = df_val_loss_avg_accur.c0_avg_accur.rolling(window=win_size).mean()

plt.plot(df_val_loss_avg_accur.steps.iloc[start:-win_size+1], df_val_loss_avg_accur.MA0[start:-win_size+1])
plt.title("C0 Avg Accuracy vs. iter_count 0 - 200000")
plt.xlabel('iter_count')
plt.ylabel('Accuracdy - Moving Avg (window=5)')
```
